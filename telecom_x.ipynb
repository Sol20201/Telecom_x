##### Distribución de evasión

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import seaborn as sns

# Contar la cantidad de clientes con abandono (churn=1) y sin abandono (churn=0)
churn_counts = df_final['abandono'].value_counts().reset_index()
churn_counts.columns = ['abandono', 'count']
churn_counts['abandono'] = churn_counts['abandono'].map({0: 'No Abandono', 1: 'Abandono'})

# Crear el gráfico de barras
plt.figure(figsize=(6, 4))
sns.barplot(x='abandono', y='count', data=churn_counts, palette='viridis')
plt.gca().spines['top'].set_visible(False)  # Oculta la línea superior
plt.gca().spines['right'].set_visible(False)  # Oculta la línea derecha
plt.title('Distribución de Clientes por Abandono', size=12, color='red')
plt.xlabel('Estado de Abandono')
plt.ylabel('Número de Clientes')

# Agregar el total de clientes a las barras
for index, row in churn_counts.iterrows():
    plt.text(row.name, row['count'] + 50, f"{row['count']}", ha='center', va='bottom', size= 8, color='grey')

plt.show()
```

##### Recuento de evasión por variables categóricas

```python
import matplotlib.pyplot as plt
# Analizar la distribución de la evasión según variables categóricas
# Seleccionar columnas categóricas (excluyendo 'id_cliente' y 'abandono')
categorical_cols = df_final.select_dtypes(include='object').columns.tolist()
categorical_cols = [col for col in categorical_cols if col not in ['id_cliente', 'género']]

# Iterar sobre cada columna categórica para visualizar la distribución de evasión
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_final, x=col, hue='abandono', palette='viridis')
    plt.title(f'Distribución de Abandono por {col.replace("_", " ").title()}')
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('Número de Clientes')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Analizar la distribución de la evasión por género
plt.figure(figsize=(6, 4))
sns.countplot(data=df_final, x='género', hue='abandono', palette='viridis')
plt.title('Distribución de Abandono por Género')
plt.xlabel('Género')
plt.ylabel('Número de Clientes')
plt.show()
```

#### Conteo de evasión por variables numéricas

```python
import matplotlib.pyplot as plt
# Seleccionar las columnas numéricas relevantes para el análisis de evasión
# Excluir 'abandono' ya que es la variable objetivo y 'ciudadano_mayor' y 'dependientes' que son binarias y ya se analizaron categóricamente
numeric_cols = ['antigüedad', 'cargos_mensuales', 'cargos_totales', 'cuentas_diarias']

# Iterar sobre cada columna numérica para visualizar su distribución por estado de abandono
for col in numeric_cols:
    plt.figure(figsize=(12, 6))
    sns.histplot(data=df_final, x=col, hue='abandono', multiple='stack', kde=True, palette='viridis')
    plt.title(f'Distribución de {col.replace("_", " ").title()} por Estado de Abandono')
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('Frecuencia')
    plt.show()

    # También es útil ver boxplots para comparar las medianas y la dispersión
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_final, x='abandono', y=col, palette='viridis')
    plt.title(f'Boxplot de {col.replace("_", " ").title()} por Estado de Abandono')
    plt.xlabel('Estado de Abandono')
    plt.ylabel(col.replace("_", " ").title())
    plt.xticks([0, 1], ['No Abandono', 'Abandono'])
    plt.show()
```

##### 🔹 Introducción
Telecom X enfrenta un problema de evasión de clientes (también llamado churn), lo que significa que una parte significativa de sus usuarios está cancelando sus servicios. El objetivo de este análisis es entender las razones detrás de esta fuga y encontrar patrones en los datos que ayuden a desarrollar estrategias para reducir la cantidad de cancelaciones.

#####🔹 Limpieza y Tratamiento de Datos
Antes de analizar los datos, realizamos una serie de pasos para organizarlos:
######✅ Se importó la información desde un repositorio en línea.
######✅ Se eliminó información duplicada y se corrigieron valores vacíos.
######✅ Se transformaron algunos datos para hacerlos más fáciles de interpretar, como convertir respuestas en Sí y No a números para facilitar el análisis.

#####🔹 Análisis Exploratorio de Datos
Para entender mejor la evasión de clientes, realizamos distintos tipos de análisis:
######📊 Distribución de Abandono: Se identificó cuántos clientes abandonaron el servicio y cuántos lo mantuvieron. Aproximadamente el 25% de los clientes han cancelado su contrato.
######📈 Relación con el tipo de contrato: Los clientes con contratos mensuales tienen una tasa de abandono significativamente mayor que aquellos con contratos de un año o más.
######📡 Impacto del servicio de internet: Los clientes con internet de fibra óptica parecen presentar una mayor tasa de abandono en comparación con aquellos que usan DSL.
######💳 Método de pago: Los usuarios que pagan con cheques electrónicos tienen un índice de cancelación más alto en comparación con otros métodos de pago.
#####🔹 Conclusiones e Insights
######🔍 Los clientes con contratos cortos tienen mayor probabilidad de cancelar el servicio.
######🔍 Los métodos de pago pueden estar relacionados con la evasión: quienes pagan con cheque electrónico tienden a cancelar más.
######🔍 La estabilidad del servicio de internet podría influir en la decisión de los clientes de permanecer o abandonar.
######🔍 Los clientes que han estado más tiempo con la empresa suelen cancelar menos, lo que sugiere que la lealtad aumenta con el tiempo.
